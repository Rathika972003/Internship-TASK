```python
plt.figure(figsize=(8, 8))  # Set the figure size for the pie chart
location_counts = df['Location'].value_counts()  # Count occurrences for each location
plt.pie(location_counts, labels=location_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))  # Create the pie chart
plt.title('Distribution of customer Locations')  # Add a title
plt.show()
```

```python
plt.figure(figsize=(10, 6))  # Set the figure size for the box plot
sns.boxplot(x='Product_Category', y='Purchase_Amount', data=df, palette='pastel')  # Create the box plot
plt.title('Product Category vs. Purchase Amount')  # Add a title
plt.xlabel('Product Category')  # Label the x-axis
plt.ylabel('Purchase Amount')  # Label the y-axis
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability if needed
plt.show()
```

```python
numeric_df = df.select_dtypes(include=['number'])

# Check correlation matrix
numeric_df.corr()
```

```python
plt.figure(figsize=(10, 6))
sns.heatmap(numeric_df.corr(), annot=True, cmap='YlGnBu')
plt.title('Correlation Heatmap')
plt.show()
```
