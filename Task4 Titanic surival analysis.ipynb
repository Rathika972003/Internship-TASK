```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C:/Users/rathi/Downloads/titanic/train.csv")
```

```python
print(data.info())
```

```python
print(data.describe())
```

```python
print("Missing Values:")
print(data.isnull().sum())
```

```python
# Plot age distribution
sns.histplot(data['Age'].dropna(), bins=20, kde=True, color='blue')
plt.title('Age Distribution of Passengers')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Categorize passengers into Age Groups
data['AgeGroup'] = pd.cut(data['Age'], bins=[0, 12, 60, 100], labels=['Child', 'Adult', 'Senior'])

# Count passengers in each group
age_group_counts = data['AgeGroup'].value_counts()
print("Passenger counts by Age Group:")
print(age_group_counts)

# Visualize Age Group counts
sns.countplot(x='AgeGroup', data=data, palette='viridis', order=['Child', 'Adult', 'Senior'])
plt.title('Passenger Count by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Count')
plt.show()
```

```python
# Bar plot for survival rate by Pclass
sns.barplot(x='Pclass', y='Survived', data=data, ci=None, palette='coolwarm')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()

# Calculate survival rate by class
survival_by_class = data.groupby('Pclass')['Survived'].mean()
print("Survival rates by Passenger Class:")
print(survival_by_class)
```

```python
# Bar plot for survival rate by gender
sns.barplot(x='Sex', y='Survived', data=data, ci=None, palette='Set2')
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.show()

# Calculate survival rate by gender
survival_by_gender = data.groupby('Sex')['Survived'].mean()
print("Survival rates by Gender:")
print(survival_by_gender)
```

```python
# Box plot to show Fare vs Survival
sns.boxplot(x='Survived', y='Fare', data=data, palette='muted')
plt.title('Fare vs Survival')
plt.xlabel('Survived')
plt.ylabel('Fare')
plt.show()

# Group fares into categories (quartiles)
data['FareGroup'] = pd.qcut(data['Fare'], 4, labels=['Low', 'Medium', 'High', 'Very High'])

# Survival rate by Fare Group
fare_survival = data.groupby('FareGroup')['Survived'].mean()
print("Survival rates by Fare Group:")
print(fare_survival)

# Bar plot for survival rates by Fare Group
sns.barplot(x='FareGroup', y='Survived', data=data, ci=None, palette='magma')
plt.title('Survival Rate by Fare Group')
plt.xlabel('Fare Group')
plt.ylabel('Survival Rate')
plt.show()
```

```python
# Check missing values in Age
print(f"Missing values in Age: {data['Age'].isnull().sum()}")

# Fill missing Age values with median
data['Age'].fillna(data['Age'].median(), inplace=True)

# Verify missing values are filled
print(f"Missing values in Age after filling: {data['Age'].isnull().sum()}")
```
